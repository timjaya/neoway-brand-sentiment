```python
import pandas as pd
import numpy as np
```

```python
reviews = pd.read_json('reviews_1500k.json', lines=True)
reviews2 = pd.read_json('reviews_1500k_2.json', lines=True)
```

```python
reviews.head()
```

```python
reviews2.head()
```

```python
business = pd.read_json('business.json', lines=True)
```

```python
business.state.unique()
```

```python
business[business.state=='ON']
```

```python
restaurants = business[business['categories'].str.contains("Restaurants", na=False)]
```

```python
restaurant_reviews = reviews[reviews.business_id.isin(restaurants.business_id)]
restaurant_reviews2 = reviews2[reviews2.business_id.isin(restaurants.business_id)]
```

```python
len(restaurant_reviews2)
```

```python
restaurants = pd.concat([restaurant_reviews, restaurant_reviews2], ignore_index=True)
```

```python
restaurants.head()
```

```python
len(restaurants)
```

```python
restaurants.to_json('restaurant_reviews_1900k.json', lines=True, orient='records')
```
